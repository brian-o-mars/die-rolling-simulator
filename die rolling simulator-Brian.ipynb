```python
import random
```

```python
print('This is a die rolling simulator')
o = input('Press "s" to start')
if o == "s":
    x = 'r'
    while x == 'r':
        no = random.randint(1,6)
        if no == 1:
            print('----------')
            print('|           |')
            print('|     0    |')
            print('|           |')
            print('|           |')
            print('----------')

        if no == 2:
            print('----------')
            print('|           |')
            print('|     0    |')
            print('|     0    |')
            print('|           |')
            print('----------')

        if no == 3:
            print('----------')
            print('|           |')
            print('|     0    |')
            print('|   0  0  |')
            print('|           |')
            print('----------')

        if no == 4:
            print('----------')
            print('|           |')
            print('|   0 0   |')
            print('|   0 0   |')
            print('|           |')
            print('----------')

        if no == 5:
            print('----------')
            print('|           |')
            print('|   0 0   |')
            print('| 0 0 0  |')
            print('|           |')
            print('----------')

        if no == 6:
            print('----------')
            print('|           |')
            print('|0  0  0 |')
            print('|0  0  0 |')
            print('|           |')
            print('----------')
        x = input('press "r" to roll again')
```
